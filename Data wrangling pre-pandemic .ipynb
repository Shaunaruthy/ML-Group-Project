```python
import os
import numpy as np 
import pandas as pd
os.getcwd()
```

```python
data = pd.read_csv("2019 Q4 BIKES.csv")
data = pd.DataFrame(data)
```

```python
data = data.drop(columns = ['LAST UPDATED','STATUS', 'ADDRESS', 'LATITUDE', 'LONGITUDE'])
data = data[data["STATION ID"].isin([69, 75, 98])]
```

```python
data.TIME = (pd.to_datetime(data.TIME, format = "%Y-%m-%d %H:%M:%S"))

data = data.groupby('STATION ID').resample('H', on= 'TIME').std()
data = data.rename(columns={'AVAILABLE BIKES': 'CHANGE'})

data
```

```python
data2 = data

data2 = data2.drop(columns ='STATION ID')
data2 = data2.reset_index()

data2.TIME = (pd.to_datetime(data2.TIME, format = "%Y-%m-%d %H:%M:%S"))


data2['Year'] = data2['TIME'].dt.strftime('%Y')
data2['Month'] = data2['TIME'].dt.strftime('%m')
data2['Day'] = data2['TIME'].dt.strftime('%d')

data2
```

```python
d2 = data2.drop(columns = ["BIKE STANDS", "AVAILABLE BIKE STANDS"])

from datetime import datetime

d2['Weekday'] = d2["TIME"].dt.day_name()
d2['Hour'] = d2["TIME"].dt.hour

d2
```

```python
os.makedirs('Desktop/MLfinal', exist_ok=True)  

d2.to_csv('2019q4NEW.csv',  index=False)
```
