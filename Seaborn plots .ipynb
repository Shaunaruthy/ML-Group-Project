```python
import os
import numpy as np 
import pandas as pd
os.getcwd()



%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Classifiers
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
#Evaluation tools
from sklearn.metrics import plot_confusion_matrix
#CV
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression




data = pd.read_csv("2020NEWCOMBINED.csv")
data = pd.DataFrame(data)
data.head()

import matplotlib.pyplot as plt
import seaborn as sns
# import Pandas libraries
import pandas as pd

sns.set_theme()


data = data[data['Month'] != 1]

data.head()

data['Month'].unique()

summary = data['CHANGE'].describe()

summary

```

```python
#lINEPLOT OF MONTH AND CHANGE

sns.lineplot(data=data, x='Month', y='CHANGE', hue = 'STATION ID')
legend = plt.legend()
legend.set_bbox_to_anchor((1,1))
legend.get_frame().set_linewidth(0.0)
plt.title("Bike Usage During the Pandemic")
plt.xlabel("Month")
plt.ylabel("Bike Usage")

plt.savefig('bikeusage_pandemic_month.png')
```

```python
#WEEK PLOT



day_order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
data['Weekday'] = pd.Categorical(data['Weekday'], categories = day_order, ordered = True)


start_date = '2020-04-06 00:00:00'
end_date= '2020-04-12 23:59:59'
data_week = data.loc[(data['TIME'] >= start_date) & (data['TIME'] <= end_date)]

data_week




sns.lineplot(data=data, x='Weekday', y='CHANGE', hue = 'STATION ID')
legend = plt.legend()
legend.set_bbox_to_anchor((1,1))
legend.get_frame().set_linewidth(0.0)
plt.title("Bike Usage in Week 2 April 2020")
plt.xlabel("Day")
plt.ylabel("Bike Usage")

plt.savefig('bikeusage_pandemic_week_april.png')
```

```python
sns.lineplot(data=data, x='Month', y='CHANGE')
plt.title("Bike Usage 2020")
plt.xlabel("Month")
plt.ylabel("Bike Usage Overall")

plt.savefig('Overall Usage 2020.png')
```
