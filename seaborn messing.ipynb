```python
data = pd.read_csv("merged_files.csv")
data = pd.DataFrame(data)
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# import Pandas libraries
import pandas as pd

sns.set_theme()

data["datetime"] = pd.to_datetime(data["TIME"])
data["month"] = data["datetime"].dt.month

#data.head()


mean = np.mean(data["Usage"])
median = np.median(data["Usage"])
std_dev = np.std(data["Usage"])
variance = np.var(data["Usage"])

variance

sns.lineplot(data=data, x='month', y='Usage')




#sns.histplot(data=data, x='Usage', bins=10, kde='true')

sns.displot(data=data, x= 'Usage', col='NAME', bins=10, col_wrap=3)

sns.catplot(kind='strip', data=data, x='NAME', y='Usage', aspect=2)

sns.catplot(data=data, kind='box', x='NAME', y='Usage', aspect=2)

sns.catplot(kind='bar',data=data,x='NAME',y='Usage',aspect=2)
```
