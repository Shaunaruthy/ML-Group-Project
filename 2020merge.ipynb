```python
import os
import numpy as np 
import pandas as pd
os.getcwd()
```

```python
data = pd.read_csv("dublinbikes_2020_secondQ.csv")
data = pd.DataFrame(data)
data.head()
```

```python
# Drop extra columns
data = data.drop(columns = ['LAST UPDATED','STATUS', 'ADDRESS', 'LATITUDE', 'LONGITUDE'])
# Select Stations
data = data[data["STATION ID"].isin([69, 75, 98])]
# Create Usage Variable by percentage of bikes NOT at the station
data["Usage"] = data["AVAILABLE BIKE STANDS"] / data['BIKE STANDS'] * 100
# Remove seconds and subset dataframe by whole hours
data['TIME'] = pd.to_datetime(data['TIME'])
data['TIME'] = data['TIME'].dt.strftime('%Y-%m-%d %H:%M')
data['TIME'] = pd.to_datetime(data['TIME'])
time = (data['TIME'].dt.minute == 0)
```
